## Exercise 2 - Stimulating the Decision Making Circuit

The input stimulus is implemented by two inhomogenous Poisson processes: The subpopulation "Left" and "Right" receive input from two different PoissonGroups (see Figure "Network Structure"). The input has a "coherence level" $c$ and is noisy. We have implemented this in the following way: every 30ms, the firing rates $\nu_{\text{left}}$ and $\nu_{\text{right}}$ of each of the two PoissonGroups are drawn from a normal distribution:

\begin{align}
\nu_{\text{left}} &\sim& \mathcal{N}(\mu_\text{left},\sigma^{2}) \\
\nu_{\text{right}} &\sim& \mathcal{N}(\mu_\text{right},\sigma^{2}) \\
\mu_{\text{left}} &=& \mu_0  (0.5 + 0.5c) \\
\mu_{\text{right}} &=& \mu_0  (0.5 - 0.5c) \\
c &\in& [-1, +1]
\end{align}

The coherence level $c$, the maximum mean $\mu_0$ and the standard deviation $\sigma$ are parameters of `sim_decision_making_network()`.

````python
def plot_decision_space(rm1, rm2, 
                        avg_window_width=100*b2.ms, 
                        label1="Population rate left [Hz]", 
                        label2="Population rate right [Hz]") -> None: 
    
    smoothed_rate1 = rm1.smooth_rate(window="flat", width=avg_window_width)/b2.Hz
    smoothed_rate2 = rm2.smooth_rate(window="flat", width=avg_window_width)/b2.Hz
    
    # Plot with time array as color code.
    plt.scatter(smoothed_rate1, smoothed_rate2, c=rm1.t / b2.ms)
    plt.xlabel(label1)
    plt.ylabel(label2)
    plt.colorbar(label="time [ms]")
    minrate = min(min(smoothed_rate1), min(smoothed_rate2))
    maxrate = max(max(smoothed_rate1), max(smoothed_rate2))
    plt.xlim([minrate, maxrate])
    plt.ylim([minrate, maxrate])
````

````python
def get_decision_time(rm1, rm2, avg_window_width=100*b2.ms, rate_threshold=20*b2.Hz):
    
    smoothed_rate1 = rm1.smooth_rate(window="flat", width=avg_window_width)/b2.Hz
    smoothed_rate2 = rm2.smooth_rate(window="flat", width=avg_window_width)/b2.Hz
    
    decision_idx1 = np.where(smoothed_rate1 > rate_threshold/b2.Hz)[0] 
    decision_idx2 = np.where(smoothed_rate2 > rate_threshold/b2.Hz)[0]

    if decision_idx1.size > 1: 
        decision_time1 = rm1.t[decision_idx1[0]] 
    else: 
        decision_time1 = 0.0 * b2.ms
        
    if decision_idx2.size > 1: 
        decision_time2 = rm2.t[decision_idx2[0]]
    else: 
        decision_time2 = 0.0 * b2.ms
    
    return decision_time1, decision_time2
````

````python
coherence_levels = [-0.1, -0.5]  # for negative values, B is the correct decision.
nr_repetitions = 3

time_to_A,\
time_to_B,\
count_A,\
count_B,\
count_No = decision_making.run_multiple_simulations(get_decision_time,
                                                    coherence_levels,
                                                    nr_repetitions,
                                                    max_sim_time=800 * b2.ms,
                                                    rate_threshold=20 * b2.Hz,
                                                    avg_window_width=100 * b2.ms)
````

````python
def percent_correct_and_rt(coherence_levels, time_to_A, time_to_B, count_A, count_B): 
    
    num_reps = time_to_A.shape[1]
    percent_correct = []
    reaction_times = []
    for idx, coh in enumerate(coherence_levels):
        if coh >= 0.: 
            pc = count_A[idx] / num_reps
        else: 
            pc = count_B[idx] / num_reps
        
        percent_correct.append(pc)
        # Take avg RT for current coherence, irrespective of direction.
        # Ignore missing decisions.
        rt = 0.5 * (time_to_A[idx,:][time_to_A[idx,:].nonzero()].mean() + 
                    time_to_B[idx,:][time_to_B[idx,:].nonzero()].mean())
        reaction_times.append(rt)    
    
    return percent_correct, reaction_times
    
````
